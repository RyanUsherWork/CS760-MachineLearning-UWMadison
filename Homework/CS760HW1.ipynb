```python
#Question 7.)
#a.)

import numpy as np
import matplotlib.pyplot as plt

mean = [1,-1]
cov = [[2,0], [0,2]]

x,y = np.random.multivariate_normal(mean,cov, 100).T
plt.plot(x,y,'x')
plt.title('Scatter Plot for Gaussian Distribution')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis('equal')
plt.show()
```

```python
mean_1 = [5,0]
cov_1 = [[1, 0.25], [0.25,1]]

mean_2 = [-5,0]
cov_2 = [[1, -.25],[-.25, 1]]

sample_1 = np.random.multivariate_normal(mean_1,cov_1,30)
sample_2 = np.random.multivariate_normal(mean_2,cov_2,70)

samples = np.vstack((sample_1,sample_2))

plt.scatter(samples[:,0], samples[:,1])
plt.title('Scatter Plot for Mixture Distribution')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```
